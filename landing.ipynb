```python


# Path to the directory containing your CSV files

root_dir ='universe_8'
#i = ["Back", "Front", "Top", "Bottom", "Left", "Right"]
# List to store DataFrames from each CSV
dfs = []

# Iterate over each CSV file in the directory
# for filename in os.listdir(root_dir):
#     if filename == "Star_Data.csv":
#         filepath = os.path.join(all_star_directories, filename)
#         # Read the CSV file into a DataFrame and append it to the list
#         dfs.append(pd.read_csv(filepath))

for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        if file == "Star_Data.csv":
            filepath = os.path.join(subdir, file)
            dfs.append(pd.read_csv(filepath))

    

# Concatenate all DataFrames in the list along the rows
combined_df = pd.concat(dfs, ignore_index=True)

# Path to the output CSV file
output_file = 'universe_8/combined_stars.csv'

# Write the combined DataFrame to a new CSV file
combined_df.to_csv(output_file, index=False)

print("Combined CSV file created successfully.")
```

```python
star_data = pd.read_csv(root_dir+'/combined_stars.csv') 
#print(stars.keys())
```

```python
import re
directions = [re.match(r'([a-zA-Z]+)', view).group() for view in star_data['Name']]

print(directions)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the star data into a DataFrame (replace 'stars.csv' with your file name)
#star_data = pd.read_csv('stars.csv')

# Create a new figure for the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define a fixed Z-coordinate for each view
view_z = {'TopS': 10, 'lefts': 20, 'fronts': 30, 'BackS': 40, 'rights': 50, 'bottoms': 60}

# Loop through each view and plot the stars
views = ['TopS', 'LeftS', 'fronts', 'BackS', 'rights', 'bottoms']
colors = ['b', 'g', 'r', 'c', 'm', 'y']  # Color for each view
for i, view in enumerate(directions):
    view_data = star_data[star_data['Name'] == view]
    ax.scatter(view_data['X'], view_data['Y'], view_z[view], c=colors[i], label=view)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Stars from Different Views')
ax.legend()

# Show the plot
plt.show()
```
